```python
import pandas as pd
```

```python
file = r"D:\Retail Pulse Project\data\online_retail_II.xlsx"

df1 = pd.read_excel(
    file,
    sheet_name="Year 2009-2010"
)

df2 = pd.read_excel(
    file,
    sheet_name="Year 2010-2011"
)

df = pd.concat(
    [df1, df2],
    ignore_index=True
)

print(df.shape)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df[
    df["Quantity"] > 0
]
```

```python
df["Revenue"] = (
    df["Quantity"]
    * df["Price"]
)
```

```python
df.isnull().sum()
```

```python
df.to_csv(
    r"D:\Retail Pulse Project\data\cleaned_retail.csv",
    index=False
)
print("Successfully Created!!")
```
