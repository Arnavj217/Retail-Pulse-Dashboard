```python
import pandas as pd
```

```python
forecast = pd.read_csv(
    r"D:\Retail Pulse Project\data\sales_forecast.csv"
)

df = pd.read_csv(
    r"D:\Retail Pulse Project\data\cleaned_retail.csv"
)
```

```python
inventory = (
    df.groupby(
        "StockCode"
    )["Quantity"]
    .sum()
    .reset_index()
)

inventory.columns = [
    "StockCode",
    "CurrentStock"
]
```

```python
forecast_demand = (
    forecast["yhat"]
    .tail(30)
    .mean()
)
```

```python
inventory["ForecastDemand"] = (
    forecast_demand
)

inventory["ReorderQuantity"] = (
    inventory["ForecastDemand"]
    - inventory["CurrentStock"]
)

inventory["ReorderQuantity"] = (
    inventory["ReorderQuantity"]
    .clip(lower=0)
)
```

```python
inventory["InventoryStatus"] = inventory[
    "ReorderQuantity"
].apply(
    lambda x:
    "Reorder"
    if x > 0
    else "Sufficient"
)
```

```python
inventory.to_csv(
    r"D:\Retail Pulse Project\data\inventory_recommendations.csv",
    index=False
)
print("Successfully Created!!")
```
